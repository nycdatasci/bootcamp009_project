```python
import pandas as pd
import numpy as np
import re
```

```python
def get_id(x):
    return re.split(pattern='id=|,', string=x)[1]

def get_title(x):
    return re.split(pattern='title=u|,', string=x)[2]

def get_idfs(x):
    return re.split(pattern="SparseVector\(256, ", string=x)[1][:-2]
```

```python
data = []
for i in range(10):
    data.append(pd.read_table('part-0000'+str(i), header=None))
for i in range(10, 74):
    data.append(pd.read_table('part-000'+str(i), header=None))
    
data = pd.concat(data).reset_index(drop=True)
```

```python
data.columns = ['raw']
data['id']=data['raw'].map(get_id)
data['title'] = data['raw'].map(get_title)
data['idfs']=data['raw'].map(get_idfs)
data.drop(labels='raw', axis=1, inplace=True)
data
```
