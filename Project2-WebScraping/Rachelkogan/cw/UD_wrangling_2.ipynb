```python
import pandas as pd
import numpy as np
import re
```

```python
ud = pd.read_csv('ud.csv')
ud = ud.dropna()
```

```python
ud['phrase']=ud['phrase'].map(str)
ud['meaning']=ud['meaning'].map(str)
ud['up_votes']=ud['up_votes'].map(int)
ud.head()
```

```python
#clean dates
ud['date'] = ud['date'].map(lambda x: re.sub(pattern='\n', repl='', string=x))

ud['month'] = ud['date'].map(lambda x: x.split(' ')[0])
ud['day'] = ud['date'].map(lambda x: x.split(' ')[1])
ud.head()
```

```python
# add length of phrase

ud['phrase_length'] = ud['phrase'].map(lambda x: len(str(x).split(' ')))
ud.head()
```

```python
ud.describe()
```

```python
ud['emoticon'] = ud['phrase'].map(lambda x: re.search('[a-zA-Z]', x) == None)
ud['emoticon'].value_counts()
```

```python
ud[ud['emoticon']==True]
```

```python

ud['acronym'] = ud.apply(lambda x: (x['phrase']==x['phrase'].upper()) & (x['emoticon']==False), axis=1)

ud['acronym'].value_counts()
ud[ud['acronym']==True]
```

```python
ud.sort_values(by='phrase_length', ascending=False)
```

```python
np.corrcoef(ud['phrase_length'], ud['up_votes']+ud['down_votes'])
```

```python
ud['phrase_length'].value_counts()
```

```python
ud['trump'] = ud['phrase'].map(lambda x: re.search('trump', x.lower()) != None)
ud['trump'].value_counts()
ud[ud['trump']==True]
```
