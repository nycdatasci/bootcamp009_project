```python
cw = pd.concat([cw_4, cw_5, cw_6, cw_7, cw_8, cw_9, cw_10, cw_11, cw_12], ignore_index=True)
print cw.shape
cw.dropna(inplace=True)
print cw.shape
cw.head()
```

```python
cw['answer'].value_counts()
vals = pd.DataFrame(cw['answer'].value_counts()).reset_index()

vals.columns = ['answer', 'answer_count']
vals.head()
```

```python
cw = pd.merge(left=cw, right=vals, how='left', on='answer')
cw['answer_count']=cw['answer_count'].map(lambda x: int(x))
cw.head()
```

```python
#find high-frequency words
freq = cw['answer'].value_counts()

high_freq_words = freq[freq.values > 30]
```

```python
cw_high_freq = cw[cw['answer'].isin(list(high_freq_words.index))]    

high_freq_dict = dict.fromkeys(list(high_freq_words.index), [])

cw_high_freq.head()
```

```python
for word in list(high_freq_words.index):
    high_freq_dict[word] = list(cw_high_freq[cw_high_freq['answer']==word]['clue'])
```

```python
for key in high_freq_dict:
    print key, high_freq_dict[key]
    print 
    print  
```

```python
oed = pd.read_csv('oed.csv')
oed.head()
```

```python
oed['entry_type']=oed['entry_type'].map(str)
oed['word']=oed['word'].map(str)
oed['month']=oed['month'].map(str)
oed['year']=oed['year'].map(str)
oed['part_of_speech']=oed['part_of_speech'].map(str)
oed.head()
```

```python
oed['cword'] = oed['word'].map(lambda x: re.sub(pattern='[ -]', repl='', string=x).upper())
oed.head()

```

```python
#find intersection of nyt cw and oed

common_words = set(oed['cword']).intersection(cw['answer'])
print len(common_words)
common_words
```
