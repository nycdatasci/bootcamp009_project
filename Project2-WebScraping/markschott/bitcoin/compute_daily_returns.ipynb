```python
### Compute daily return value for bitcoin
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#dataset = 'data/btceUSD.csv'
#df = pd.read_csv(dataset, names = ['timestamp','price','amount'])
dataset = 'data/daily_price_history.csv'
df = pd.read_csv(dataset, index_col=0)
```

```python
df
```

```python
plt.plot(df.timestamp,df.price)
```

```python
daily_return = [(df.price[i] - df.price[i-1])*100/df.price[i-1] for i in range(1,len(df.price))]
daily_return
```
