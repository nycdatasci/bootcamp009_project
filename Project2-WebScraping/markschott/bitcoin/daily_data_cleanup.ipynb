```python
### Clean up data fetched from 99bitcoins.com/price-chart-history (Take 2)
import numpy as np
import pandas as pd
```

```python
line = []
with open('bitcoin_prices_by_data.txt','r') as f:
    line = f.read()
```

```python
## Remove all brackets
line = line.replace('[','')
line = line.replace(']','')
```

```python
## Now split it on comma
data = line.split(',')
```

```python
print len(data)
print data[0]
print data[len(data)-1]
```

```python
### Now create two lists, one for the timestamp and one for the values,
### How do I convert the 13 digit posix timestamp to a normal date time (look at cell above)
timestamps = [int(data[i]) for i in range(0,len(data),2)]
prices = [float(data[i]) for i in range(1,len(data),2)]
print len(timestamps)
print len(prices)
print type(timestamps[0])
print type(prices[0])
```

```python
ts = pd.DataFrame([timestamps,prices])
ts = ts.T
ts.columns=['timestamp','price']
ts
```
