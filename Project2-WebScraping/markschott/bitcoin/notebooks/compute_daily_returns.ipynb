```python
"""
Create my own cauchy distribution to fit to the data using scipy.stats so I can atleast 
retrieve the fitting parameters which isn't done in the plot above (or at least I don't know how).

A drawback is that this method still doesn't retrieve the fit quality but scipy has a different module to 
do that called scipy.odr
"""

fig, ax = plt.subplots(figsize=(14,8))
 
ax.set_xlabel('Daily Return Rate (DRR)',fontsize=16)
ax.set_title(r'Bitcoin Daily Return Rate PDF', fontsize=20)
plt.ylim((0,0.35))
plt.xlim((-20,20))

# grid labels y=Constant
for ymaj in ax.yaxis.get_majorticklocs():
    ax.axhline(y=ymaj,ls='-', linewidth=0.4)
    
loc,scale = cauchy.fit(daily_return)
mu,sigma = norm.fit(daily_return)
xdom = np.linspace(-40,40,1000)
distcauchy = cauchy.pdf(xdom,loc,scale)
distnorm = norm.pdf(xdom,mu,sigma)

ax.hist(daily_return, bins = 200, normed=True)


ax.plot(xdom,distcauchy,label='Cauchy Distribution\nMedian={d}\nScale={e}'.format(d=round(loc,3),e=round(scale,3)))
ax.plot(xdom,distnorm,label="\nNormal Distribution\nMu={d}\nSigma={e}".format(d=round(mu,3),e=round(sigma,3)))

ax.legend(loc=0,fontsize='large',fancybox=True, shadow=True)
ax.set_facecolor('white')

fig.savefig("bitcoin_daily_return.svg",format='svg')

plt.show()
```

```python
### Look at the bitcoin price return over varying time periods
### Above was by one day but do by increments of 5
from operator import add
return_rates = []
increments = [5*i for i in range(1,10)]
for lag in increments:
    #print lag
    tmp = [0]*(len(df.price)-lag)
    for t in range(len(df.price)-lag):
        tmp[t] = (df.price[t+lag] - df.price[t])*100/df.price[t]
    #flatten the list
    #tmp = reduce(add,tmp)
    return_rates.append(tmp)
```

```python
from math import sqrt, floor
from scipy.stats import t,f

xdom = np.linspace(-40,40,1000)
    
def plot_hists(num_of_plots):
    x = float(sqrt(num_of_plots))
    
    if x != floor(x):
        print 'make plot number a rational square root'
        return
    
    if num_of_plots > len(return_rates):
         print 'Warning: number of plots greater than list size. Ending now'
         return
        
    plt.figure(2, figsize=(18,18))
    bins_no = 150
    normal = True
    count = int(100*x + 10*x + 1)
    
    for i in range(num_of_plots):
        plt.subplot(count)
        
        if i < len(return_rates):
            loc,scale = cauchy.fit(return_rates[i])
            distcauchy = cauchy.pdf(xdom,loc,scale)
            plt.hist(return_rates[i], bins = bins_no, normed = normal)
            plt.plot(xdom, distcauchy)
            
        count = count + 1
        
    plt.show()

plot_hists(4)
```
